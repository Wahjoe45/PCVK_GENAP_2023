```python
# Membuat bentuk geometri 2D dari OpenCV
# Membuat black image dengan type data int16
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
# Perhatikan koordinat titik pt1 dan pt2
# membuat persegi panjang berwarna hijau koordinat pt1 dan pt2
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

```python
# membuat persegi panjang berwarna biru sesuai koordinat pt1 dan pt2
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

```python
# membuat lingkaran sesuai radius yang tertulis pada kode
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

```python
# menambah garis sesuai koordinat pt1 dan pt2
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

```python
# membuat lingkaran sesuai radius dan koordinat yang tertulis pada kode
cv.circle(black_img,center=(400,400),radius=10,color=(255,0,0),thickness=100)
plt.imshow(black_img)
```

```python
# Pembuatan kalimat string pada gambar
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
# pembuatan ulang black image dengan type data int32
black_img2 = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

```python
# Inisialisai NumPY array dengan tipe data int32 
vertices = np.array([[100,300], [200,200], [400,300], [200,400]],dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2))
# nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yang mewakili R,G,B
pts
```

```python
# Penambahan polyline pada black image kedua
cv.polylines(black_img2,[pts],isClosed=True, color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```
